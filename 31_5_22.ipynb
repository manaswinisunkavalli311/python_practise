# pie charts 

```python
#1
import matplotlib.pyplot as plt
import numpy as np
y = np.array([30, 20, 15, 45])
plt.pie(y)
plt.show() 
```

```python
#2 labelling
import matplotlib.pyplot as plt
import numpy as np
y = np.array([30, 20, 35, 45])
mylabels = ["hyd", "vizag", "banglore", "Delhi"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#3 angle
import matplotlib.pyplot as plt
import numpy as np
y = np.array([5, 15, 45, 35])
mylabels = ["delhi", "goa", "agra", "jk"]
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
#4 labels and explode
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 15, 45])
mylabels = ["delhi", "goa", "agra", "jk"]
myexplode = [0.5, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
#5 shadow
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30,30,20,20])
mylabels = ["delhi", "goa", "agra", "jk"]
myexplode = [0.4, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
#6 colors
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30,40,10,10])
mylabels = ["delhi", "goa", "agra", "jk"]
mycolors = ["black", "pink", "b", "g"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
#7 legend
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30,14,20,50])
mylabels = ["delhi", "goa", "agra", "jk"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
#8 title 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([40, 10, 16, 34])
mylabels = ["delhi", "goa", "agra", "jk"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```

```python
#1 Add labels to the x- and y-axis:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([ 130, 100, 105, 120, 125])
y = np.array([ 20, 30, 40, 30, 60])
plt.plot(x, y)
plt.xlabel("Average height")
plt.ylabel("bmi")
plt.show()
```

```python
#3Set font properties for the title and labels:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([70, 80, 90, 100])
y = np.array([240, 200, 160, 300])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':10}
plt.title("body Data", fontdict = font1)
plt.xlabel("Average  height", fontdict = font2)
plt.ylabel("bmi", fontdict = font2)
plt.plot(x, y)
plt.show()
```

# grids

```python
#1 Add grid lines to the plot:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80, 85, 90, 95, 100])
y = np.array([100,200,300,400,500])
plt.title("Data")
plt.xlabel("Averageheight")
plt.ylabel("bmi")
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#2Display only grid lines for the x-axis:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([60, 85, 90, 75, 125])
y = np.array([ 290, 300, 310, 400, 430])
plt.title("Data")
plt.xlabel("Average height")
plt.ylabel("bmi")
plt.plot(x, y)
plt.grid(axis = 'x')
plt.show()
```

```python
#3 Display only grid lines for the y-axis:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80,55,65,75])
y = np.array([230,250,400,500])
plt.title(" Data")
plt.xlabel("Average height")
plt.ylabel("bmi")
plt.plot(x, y)
plt.grid(axis = 'y')
plt.show()
```

```python
#4 Set the line properties of the grid:
import numpy as np
import matplotlib.pyplot as plt
x = np.array([ 95, 100, 105, 110, 125])
y = np.array([100,200,300,400,500])
plt.title("data")
plt.xlabel("Average height")
plt.ylabel("bmi")
plt.plot(x, y)
plt.grid(color = 'red', linestyle = ':', linewidth = 0.7)
plt.show()
```
